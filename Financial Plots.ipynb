```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import cufflinks as cf
import plotly.offline as plyo
```

```python
plt.style.use('seaborn')
mpl.rcParams['font.family'] = 'serif'
sns.set()
plyo.init_notebook_mode(connected=True)
```

```python
raw = pd.read_csv('/Users/andy/Desktop/py4fi2nd-master/source/fxcm_eur_usd_eod_data.csv',
                 index_col=0, parse_dates=True)
```

```python
raw.info()
```

```python
quotes = raw[['AskOpen', 'AskHigh', 'AskLow', 'AskClose']]
quotes = quotes.iloc[-60:]  # Only a few data rows are used for the visualization
quotes.tail()
```

```python
qf = cf.QuantFig(
    quotes,  # The DataFrame object is passed to the QuantFig constructor
    title='EUR/USD Exchange Rate',
    legend='top', name='EUR/USD')
```

```python
plyo.iplot(
    qf.iplot(asFigure=True),
# image='png',
filename='qf_01'
)
```

```python
qf.add_bollinger_bands(periods=15,  # the number of periods for the Bollinger band
                      boll_std=2)  # the number of standard deviations to be used for the band width
```

```python
plyo.iplot(qf.iplot(asFigure=True),
          # image='png',
          filename='qf_02'
          )
```

```python
qf.add_rsi(periods=14,  # fixes the RSI period
          showbands=False)  # does not show an uppper or lower band
```

```python
plyo.iplot(qf.iplot(asFigure=True),
          # image='png',
          filename='qf_03')
```
