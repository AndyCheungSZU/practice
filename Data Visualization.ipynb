```python
def func(x):
    return 0.5 * np.exp(x) + 1
a, b = 0.5, 1.5
x = np.linspace(0, 2)
# The x values to plot the function
y = func(x)
# The y values to plot the function
Ix = np.linspace(a, b)
# The x values within the integral limits
Iy = func(Ix)
# The y values within the integral limits
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
# The list object with multiple tuple objects representing coordinates for the polygon to be plotted.
```

```python
from matplotlib.patches import Polygon

fig, ax = plt.subplots(figsize=(10, 6))
plt.plot(x, y, 'b', lw=2)  # Plots the function values as a blue line
plt.ylim(bottom=0)  # Defines the minimum y value for the ordinate axis

# Plots the polygon (integral area) in gray
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

# Places the integral formula in the plot
plt.text(0.5 * (a + b), 1, r'$\int_a^b f(x)\mathrm{d}x$',
        horizontalalignment='center', fontsize=20)

# Places the axis labels
plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(x)$', '$f(b)$'))
```

```python
strike = np.linspace(50, 150, 24)
```

```python
ttm = np.linspace(0.5, 2.5, 24)
```

```python
strike, ttm = np.meshgrid(strike, ttm)
# The two two-dimensional ndarray objects (grids) created
```

```python
strike[:2].round(1)
```

```python
iv = (strike - 100) ** 2 / (100 * strike) / ttm
```

```python
iv[:5, :3]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10, 6))
ax = fig.gca(projection='3d')  # Sets up a canvas for 3D plotting

# Creates the 3D plot
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                      cmap=plt.cm.coolwarm, lw=0.5, antialiased=True) 
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5)
```

```python
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.view_init(30, 60)  # Sets the viewing angle
ax.scatter(strike, ttm, iv, zdir='z', s=25,
          c='b', marker='^')
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
```

```python
import pandas as pd
import cufflinks as cf
import plotly.offline as plyo

plyo.init_notebook_mode(connected=True)
# turns on the notebook plotting mode
```

```python
import numpy as np

a = np.random.standard_normal((250, 5)).cumsum(axis=0)
```

```python
index = pd.date_range('2019-1-1',
                     freq='B',  # The frequency (“business daily”)
                     periods=len(a))  # The number of periods needed
```

```python
df = pd.DataFrame(100 + 5 * a,
                 columns=list('abcde'),
                 index=index)
```

```python
df.head()
```

```python
plyo.iplot(df.iplot(asFigure=True),
          # image='png',
          filename='ply_01')
# parameter asFigure=True to allow for local plotting and embedding
# the image option provides in addition a static bitmap version of the plot
```

```python
plyo.iplot(
    df[['a', 'b']].iplot(asFigure=True,
                        theme='polar',  # selects a theme(plotting style) for the plot
                        title='A Time Series Plot',
                        xTitle='date',
                        yTitle='value',
                        mode={'a': 'markers', 'b': 'lines+markers'},  # defines the plotting mode(line, marker, etc.)
                        symbol={'a': 'circle', 'b': 'diamond'},  # defines the symbols to be used as markers by column
                        size=3.5,  # fixes the size for all markers
                        color={'a': '#99ffff', 'b': 'magenta'},
                        ),
# image='png',
filename='ply_02'
)
```

```python
plyo.iplot(
    df.iplot(kind='hist',  # specifies the plotting style
            subplots=True,  # requires separate subplots for every columns
            bins=15,  # sets the bins parameter(buckets to be used = bars to be plotted)
            asFigure=True),
# image='png',
filename='ply_03'
)
```
